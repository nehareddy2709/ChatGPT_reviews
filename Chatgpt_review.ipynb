```python
# For userName, reviewCreatedVersion, and appVersion, we'll replace missing values with a placeholder
data['userName'].fillna("Unknown", inplace=True)
data['reviewCreatedVersion'].fillna("Not specified", inplace=True)
data['appVersion'].fillna("Not specified", inplace=True)

# For content, since it's a text column, we'll replace missing values with an empty string
data['content'].fillna("", inplace=True)
```

Text Normalization


```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud,STOPWORDS
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.tokenize.toktok import ToktokTokenizer
from bs4 import BeautifulSoup
import re
```

```python
stopwords = nltk.corpus.stopwords.words('english')

```

```python
def noiseremoval_text(text):
  soup = BeautifulSoup(text, "html.parser")
  text = soup.get_text()
  text = re.sub('\[[^]]*\]','',text)
  return text
```

```python
data['content'] = data['content'].apply(noiseremoval_text)
```

```python
data.head()
```

Stemming

```python
def stemmer(text):
  ps = nltk.porter.PorterStemmer()
  text = ' '.join([ps.stem(word) for word in text.split()])
  return text
```

```python
data['content'] = data['content'].apply(stemmer)
```

```python
data.head()
```

stop Words

```python
def removing_stopwords(text, is_lower_case= False):
  tokenizer = ToktokTokenizer()
  tokens = tokenizer.tokenize(text)
  tokens = [i.strip() for i in tokens]
  if is_lower_case:
    filtokens = [i for i in tokens if token not in stopwords]
  else:
    filtokens = [i for i in tokens if i.lower() not in stopwords]
  filtered_texts = ' '.join(filtokens)
  return filtered_texts
```

```python
data['content'] = data['content'].apply(removing_stopwords)
```

```python
data.head()
```

Train and test the data

```python
train_chatgpt_reviews_data = data.content[:60000]
```

```python
test_chatgpt_review_data = data.content[60000:]
```

Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(min_df = 0, max_df = 1, binary = False, ngram_range=(1,3))
cv_train = cv.fit_transform(train_chatgpt_reviews_data)
cv_test = cv.fit_transform(test_chatgpt_review_data)

print("Bag of Words for training data:", cv_train.shape)
print("Bag of Words for testing data:", cv_test.shape)
```

TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(min_df = 0, max_df = 1, use_idf = True, ngram_range = (1,3))
tf_train = tf.fit_transform(train_chatgpt_reviews_data)
tf_test = tf.fit_transform(test_chatgpt_review_data)
print("TF- IDF training data:", tf_train.shape)
print("TF-IDF testing data:", tf_test.shape)
```
